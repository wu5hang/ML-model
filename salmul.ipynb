```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('salmul_data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_before = regressor.predict(X_test)
```

```python
y_pred_new = regressor.predict([[5, 2, 1, 2]])
print("Predicted salary is for x_new is: ", y_pred_new)
```

```python
accuracy = regressor.score(X_test, y_test)
print("Accuracy of model: ", accuracy)
```

```python
plt.scatter(range(len(y_test)), y_test, color='red', label='Actual salary')
plt.scatter(range(len(y_test)), y_pred_before, color='blue', label='Predicted Salary (Before Training)')
plt.hlines(y_pred_new, xmin=0, xmax=len(y_test), color='green', linestyles='dashed', label='Predicted salary for new employee')
plt.title('Actual vs predicted salary (before training)')
plt.xlabel('Employee index')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
