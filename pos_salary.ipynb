```python
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# importing the dataset
dataset = pd.read_csv("pos_salaries.csv")
X = dataset.iloc[ :  , 1 : -1].values
y = dataset.iloc[ :  , -1].values
```

```python
# training the linear regression model on whole dataset
from sklearn.linear_model import LinearRegression
lr1 = LinearRegression()
lr1.fit(X, y)
```

```python
# training the polynomial regression on whole dataset
from sklearn.preprocessing import PolynomialFeatures

polyreg = PolynomialFeatures(degree=4)
X_poly = polyreg.fit_transform(X)
lr2 = LinearRegression()
lr2.fit(X_poly, y)
```

```python
# Visualising the Linear Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, lr1.predict(X), color = 'blue')

plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# visualizing polynomial regression result
plt.scatter(X, y, color = 'red')
plt.plot(X, lr2.predict(X_poly), color = 'blue')
plt.title('Truth or bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
